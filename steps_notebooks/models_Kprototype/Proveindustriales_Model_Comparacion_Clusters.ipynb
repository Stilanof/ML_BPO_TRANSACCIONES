# Comparing Clusters

We'll compare diffent model's clusters using *Factor Analysis of Mixed Data*. The idea is to choose the best performer: more defined and balanced clusters.

```python
#df = pd.read_csv("clusters_data.csv")

#pip install prince

from prince import FAMD
import pandas as pd
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

### Chossing k by Kprototype

```python


data_collect=["Kprototype_clusters_2_Estado.csv","Kprototype_clusters_3_Estado.csv", "Kprototype_clusters_2:Companhia.csv", "Kprototype_clusters_3:Compahnia.csv", "Kprototype_clusters_2_Categoria.csv","Kprototype_clusters_3_Categoria.csv","Kprototype_clusters_sinTopPerfomers.csv","Kprototype_clusters_sinTopPerfomers_3.csv"]


for i in data_collect:
    X=pd.read_csv(i)

    famd = FAMD(n_components =2, n_iter = 3, random_state = 101)

    famd.fit(X)
    famd.transform(X)
    famd.plot_row_coordinates(X,figsize=(15, 10),color_labels=['Clusters : {}'.format(t) for t in X['Segment']])
    
```
